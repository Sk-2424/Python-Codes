<b>Given a DataFrame with sales data, pivot the DataFrame to show the sales of each product for each month.</b>

```python
import pandas as pd

data = {'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
        'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Sales': [100, 150, 120, 180, 90, 110]}

df = pd.DataFrame(data)
df
```

```python
df.pivot(index='Month',columns='Product',values='Sales')
```

<b>Given a DataFrame with temperature data, pivot the DataFrame to show the average temperature and humidity for each city and each month.</b>

```python
import pandas as pd

data = {'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
        'City': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Temperature': [20, 25, 22, 28, 18, 23],
        'Humidity': [40, 50, 35, 45, 30, 40]}

df = pd.DataFrame(data)
df
```

```python
df.pivot(index='Month',columns='City',values=['Temperature','Humidity'])
```

<b>Given a DataFrame with sales data where each product can have multiple entries for a month, pivot the DataFrame to show the total sales for each product in each month.</b>

```python
import pandas as pd

data = {'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
        'Product': ['A', 'A', 'B', 'B', 'A', 'A'],
        'Sales': [100, 120, 80, 90, 110, 95]}

df = pd.DataFrame(data)
df
```

```python
dt=df.groupby(['Month','Product']).sum()
```

```python
dt.reset_index(inplace=True)
```

```python
dt.pivot(index='Month',columns='Product',values='Sales')
```

```python
df.pivot_table(index='Month',columns='Product',values='Sales',aggfunc='sum',fill_value=0,sort=False)
```

<b>Given a DataFrame with temperature data, pivot the DataFrame to show the maximum temperature and average humidity for each city.</b>

```python
import pandas as pd

data = {'City': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Temperature': [25, 28, 22, 30, 18, 20],
        'Humidity': [50, 45, 40, 55, 30, 35]}

df = pd.DataFrame(data)
df
```

```python
df.pivot_table(index='City',values=['Temperature','Humidity'],aggfunc={'Temperature':'max','Humidity':'mean'},sort=False)
```

<b>Given a DataFrame with information about sales and expenses for different products, use the melt function to transform the DataFrame, keeping the 'Date' column as an identifier variable.</b>

```python
import pandas as pd

data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03'],
        'Product_A_Sales': [100, 120, 80],
        'Product_B_Sales': [80, 90, 110],
        'Product_A_Expenses': [30, 40, 25],
        'Product_B_Expenses': [20, 35, 30]}

df = pd.DataFrame(data)
df
```
